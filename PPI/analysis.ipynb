```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PPIACO.csv')
df['observation_date'] = pd.to_datetime(df['observation_date'])
# for column in ['Price', 'Open', 'High', 'Low']:
#     df[column] = df[column].str.replace(',', '').astype(float)
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df["observation_date"], df["PPIACO"], linewidth=1.5)

plt.title("PPIACO Over Time (Monthly)")
plt.xlabel("Date")
plt.ylabel("Producer Price Index by Commodity: All Commodities")
# plt.legend(title="Skin Name", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()
```
