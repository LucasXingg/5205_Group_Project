```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('S&P 500 Historical Data.csv')
df['Date'] = pd.to_datetime(df['Date'])
for column in ['Price', 'Open', 'High', 'Low']:
    df[column] = df[column].str.replace(',', '').astype(float)
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df["Date"], df["Price"], linewidth=1.5)

plt.title("Close Price of S&P 500 Over Time")
plt.xlabel("Date")
plt.ylabel("Price (Close)")
# plt.legend(title="Skin Name", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()
```
