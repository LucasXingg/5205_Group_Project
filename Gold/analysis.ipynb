```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('XAU_1Month_data.csv', sep=';')
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df["Date"], df["Close"], linewidth=1.5)

plt.title("Gold Price Over Time (Monthly)")
plt.xlabel("Date")
plt.ylabel("Price (Close) (USD)")
# plt.legend(title="Skin Name", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()
```
